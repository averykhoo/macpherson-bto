```python
from bs4 import BeautifulSoup
import csv
```

```python
with open('UNIT_2021-05_BTO_R0xfTjFDMTNfMTYxOTUwMjc4NDU1Ng.xml') as f:
    soup = BeautifulSoup(f.read(), 'lxml-xml')
```

```python
rows = []
for block in soup.find_all('block'):
    block_number = block.find('block-number')
    for flat_type in block.find_all('flat-type'):
        _type = flat_type.find('type')
        for level in flat_type.find_all('level'):
            level_number = level.find('level-number')
            print(block_number.text, _type.text, level_number.text)
            for unit in level.find_all('unit'):
                unit_id = unit.find('unit-id')
                unit_number = unit.find('unit-number')
                area_sqm = unit.find('area')
                price = unit.find('price')
                print(unit_id.text,
                      unit_number.text,
                      area_sqm.text,
                      price.text)
                rows.append([block_number.text, _type.text, f'#{level_number.text}',
                            unit_id.text,
                            unit_number.text,
                            area_sqm.text,
                            price.text
                           ])
```

```python
with open('tmp.csv', 'w', newline='') as f:
    c = csv.writer(f)
    c.writerow(['block_number', 'flat_type', 'level_number', 'unit_id', 'unit_number', 'area_sqm', 'price'])
    c.writerows(rows)
```
